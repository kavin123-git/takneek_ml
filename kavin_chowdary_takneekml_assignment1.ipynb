```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#function to check if present in adjacent cells and distance formula
def dist(r,dim):
  req_pts = []  # will store points satisfying condition
  points = [[np.random.rand() * 100 for i in range(dim)] for j in range(1000)]
  grid_points = [[int(points[i][j] // r) for j in range(dim)] for i in range(1000)]   #grid index wrt r
  change = [[[abs(grid_points[i][k] - grid_points[j][k]) for k in range(dim)] for j in range(i + 1, 1000)] for i in range(1000)] # Calculating the change in grid number
 #k represents particular dimension axis

  for i in range(1000):           #checking for all points
    for j in range(i + 1, 1000):
      check = True
      for k in range(dim):     #fixing particular dimension to check
        if(change[i][j - i - 1][k] != 1 and change[i][j - i - 1][k] != 0):
            check = False
            break
      if check == True:  # checking distance
        dist = 0
        for k in range(dim):
          dist += (points[i][k] - points[j][k]) ** 2                #distance formula application
        dist = math.sqrt(dist)
        if dist < r:
            req_pts.append((points[i], points[j], dist))

  return req_pts
```

```python
# Set the seed for reproducibility
np.random.seed(42)

# Generate 1000 random RGB colors
random_colors = np.random.rand(1000, 3)

#random points btw 0 to 100
x = np.random.rand(1000) * 100
y = np.random.rand(1000) * 100

#coordinate axis
for r in range(1, 11):
  fig = plt.figure()
  axis_fig = fig.add_subplot(1, 1, 1)

  axis_fig.grid(which='both')
  xgrid = np.arange(0, 101, r)
  ygrid = np.arange(0, 101, r)

  axis_fig.grid(alpha=1, color = "black")
  axis_fig.set_xticks(xgrid)
  axis_fig.set_yticks(ygrid)

  plt.xlabel("x axis")
  plt.ylabel("y axis")
  plt.title("%sx%s table" %(r, r))
  plt.scatter(x, y, c = random_colors)

  plt.show()
  print('')
```

```python
r = eval(input("Enter the desired grid size: "))
dim= int(input("Enter dimension : "))
final_pts = pd.DataFrame(dist(r,  dim), columns = ["1st point", "2nd point", "Distance between points"])
final_pts
```
